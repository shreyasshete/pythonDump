<br>1.  An example file called stock.txt is included in the homework files.
<br>The file contains daily headline of Apple inc.’s stock price and movement. You are required to write a program that reads the file, extract the date and stock price from each day’s headline, and list them below.
<br>
<br>Please use consistent format throughout your report. Including the $ before price, and transform date into month/dd format (e.g. May 2 -> May 02)

```python
import re

fhand = open('stock.txt') #Opening File
inputString = ''
dateRegex = r'[A-Z][a-z][a-z] [0-9][0-9]?' #Regex to extract MMM dd 
amountRegex = r'[0-9][0-9][0-9]?.[0-9]{2}' # Regex to extract amount
for line in fhand:
    inputString = inputString + line #Converting file to a string
    
dateList = re.findall(dateRegex, inputString) # Regex passed to extract the required date
amountList = re.findall(amountRegex, inputString) # Regex passed to extract the required amount

fdateList = []


for element in dateList:
    fpart = ''
    spart = ''
    if len(element.split(' ')[1]) == 1: #Check if the date is d we expect it to be dd
        fpart = element.split(' ')[0] #Split to get MMM
        spart = element.split(' ')[1] #Split to get d
        fdateList.append(f'{fpart} 0{spart}') #padding zero for the dates that are not in dd format and adding to date list
    else:
        fdateList.append(element) # adding element to date list



print('Apple inc. stock prices:')
j=0
for i in range(len(dateList)):
    print(f'{fdateList[i]}, ${amountList[j]}') #Display Output
    j += 1
```
